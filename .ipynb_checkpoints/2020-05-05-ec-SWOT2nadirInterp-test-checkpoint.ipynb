```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

from scipy import interpolate
```

## Input data

In the cell below, we create a 2D pressure field on a regular grid. We also simulate the location of measurements along a line that crosses the domain.

```python
# A regular 2D grid
Nx, Ny = 100, 80    # Number of grid points in the domain
x=np.arange(0,Nx+1)*10/Nx    # Zonal coordinate
y=np.arange(0,Ny+1)*8/Ny     # Meridional coordinate
dx, dy = x[1]-x[0], y[1]-y[0]   # grid mesh size (regular)

# Define some pressure field on this grid
def make_pressure(x,y):
    """Create a 2D field with analytical functions"""
    a = 0.6*(x-4)**2 + 0.3*(y-3)**2
    b = 0.45*(x-8)**2 + 0.55*(y-5)**2
    d = 0.2*(x-2.5)**2 + 0.9*(y-6)**2
    c = np.exp(-a/2)-0.5*np.exp(-b/2)-0.6*np.exp(-d)
    c /= np.max(c)
    return c

ym, xm = np.meshgrid(y,x)
pressure = make_pressure(xm,ym)

# Nadir observation points
xn = np.arange(0.1,Nx,3)*10/Nx/np.cos(30*np.pi/180)    # Zonal coordinate
yn = 0.3 * xn + 2

# Plot
plt.contour(xm, ym,pressure)
plt.scatter(xn,yn)
```

## Interpolation

We use the function RectBivariateSpline because the grids are regular. See the doc at
https://docs.scipy.org/doc/scipy/reference/generated/scipy.interpolate.interp2d.html#scipy.interpolate.interp2d
This is the doc for the interp2d function. In there, it is suggested to use RectBivariateSpline if the grids are regular. It is much faster. The doc is more extended for interp2d.

We first create the interpolation function:

```python
swot2nadir_interp = interpolate.RectBivariateSpline(x, y, pressure)
```

Then apply this function to the nadir coordinates:

```python
pn2d = swot2nadir_interp(xn,yn)
```

We note though that the resulting field is not along a line, it is a 2D field:

```python
pn2d.shape
```

This is because RectBivariateSpline interpolates from a 2D grid to another 2D grid, assuming that inputs xn and yn define a regular grid. So, to get the values along our nadir line, we must take the diagonal of this array: 

```python
pn = np.diag(pn2d)
```

Now we plot the original pressure field and the interpolated one:

```python
plt.figure(figsize=(10,8))
plt.contour(xm, ym, pressure)
#plt.scatter(xn_m, yn_m, s = 2, c=pressure_n)
plt.scatter(xn, yn, s = 4, c=pn)
```

```python
plt.figure(figsize=(10,8))
plt.contour(xm, ym, pressure)
plt.scatter(xm, ym, s = 25, c=pressure)
plt.scatter(xn, yn, s = 4, c=pn)
```
